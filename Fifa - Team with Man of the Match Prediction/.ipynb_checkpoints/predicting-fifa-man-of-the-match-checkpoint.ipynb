# Problem Statement :

## The following data has been collected from 2018 FIFA World Cup Russia Official App. Our goal is to predict the team which will win the 'Man of the Match' Award after the Final Match.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier
import tensorflow as tf
```

```python
df = pd.read_csv("FIFA 2018 dataset.csv")
df.head()
```

```python
df.shape
```

```python
data = df.copy()
data.info()
```

# Preprocessing

```python
# Date is just a unique identifier here and doesn't help us in finding similarities within the dataset
# Hence it is safe to drop the 'Date' column

data.drop('Date', axis=1, inplace=True)
```

## Dealing With Missing Values

```python
data.isnull().sum()
```

### Features : 'Own goals' & 'Own goal Time' contain (116*100)/128 = 90.625% Missing values 

### and hence these 2 features need to be dropped

```python
data.drop(['Own goals', 'Own goal Time'], axis=1, inplace=True)
```

### Filling the NaN values from '1st Goal' feature with its mean value

```python
data['1st Goal'] = data['1st Goal'].fillna(data['1st Goal'].mean())
```

### No null values left within the dataset

## Exploratory Data Analysis (EDA)

## Correlation Heatmap

```python
corr = data.corr()

plt.figure(figsize=(20, 8))
sns.heatmap(corr, annot=True, cmap='RdYlGn')
plt.show()
```

## Correlated Features :

1) 'PSO' & 'Goals in PSO' (0.98) : drop = 'PSO' (-4.9e-18 correlated with the O/p feature 'Man of the Match')

2) 'Goals in PSO' & 'Distance Covered (Kms)' (0.72) : drop = 'Distance Covered (Kms)' (0.015 correlated with the O/p feature 'Man of the Match')

3) 'Passes' & 'Ball Possession %' (0.88) : drop = 'Passes' (0.1 correlated with the O/p feature 'Man of the Match')

4) 'Attempts' & 'On-Target' (0.73) : drop 'Attempts' (0.18 correlated with the O/p feature 'Man of the Match')

```python
data.drop(['PSO', 'Distance Covered (Kms)', 'Passes', 'Attempts'], axis=1, inplace=True)
```

```python
data.head()
```

## Outliers

```python
plt.figure(figsize=(25, 10))

data.boxplot()
plt.show()
```

### There aren't much outliers present within the dataset

## Encoding

```python
data.dtypes
```

```python
print(f"Team: {data['Team'].unique()}\n")
print(f"Opponent: {data['Opponent'].unique()}\n")
print(f"Man of the Match: {data['Man of the Match'].unique()}\n")
print(f"Round: {data['Round'].unique()}\n")
```

## Label Encoding : 

## Due to hierarchy (order) present in the features

```python
label_encoder = LabelEncoder()

data['Man of the Match'] = label_encoder.fit_transform(data['Man of the Match'])
man_mappings = {index: label for index, label in enumerate(label_encoder.classes_)}
print('Man of the Match mapping : \n', man_mappings)
```

```python
data.head()
```

```python
round_values = list(data['Round'].unique())
round_values
```

```python
round_mappings = {label: index for index, label in enumerate(round_values)}
round_mappings
```

```python
data['Round'] = data['Round'].apply(lambda x: round_mappings[x])
```

```python
data['Round']
```

## OneHotEncoding for the features : 'Team' & 'Opponent'

## as there is no order relationship present within these features

```python
data['Team'].unique()
```

```python
data['Opponent'].unique()
```

```python
pd.get_dummies(data['Team'])
```

```python
# Differentiating from the 'Team' variable

pd.get_dummies(data['Opponent'].apply(lambda x: "opp_" + x))
```

```python
data['Opponent'] = data['Opponent'].apply(lambda x: "opp_" + x)
```

```python
data
```

```python
# Concatenating the encoded 'Team' & 'Opponent' variables to the dataset

data_concat = pd.concat([data, pd.get_dummies(data['Team']), pd.get_dummies(data['Opponent'])], axis=1)
```

```python
# Dropping the original 'Team' & 'Opponent' variables from the dataset

data_concat.drop(['Team', 'Opponent'], axis=1, inplace=True)
```

```python
data_concat.head()
```

```python
# No text values within the dataset

np.sum(data_concat.dtypes == 'object')
```

## Train - Test Split

```python
y = data_concat['Man of the Match']
X = data_concat.drop('Man of the Match', axis=1)
```

```python
'Man of the Match' in X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
```

## Feature Scaling : Standardizing

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
# Standardized Training Data
X_train = pd.DataFrame(X_train, columns=X.columns)
X_train.head()
```

```python
# Standardized Testing Data
X_test = pd.DataFrame(X_test, columns=X.columns)
X_test.head()
```

# Training

```python
# Using Random Forest Model

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
# Using Artificial Neural Network(ANN) Model

inputs = tf.keras.Input(shape=(81,))
x = tf.keras.layers.Dense(128, activation=tf.nn.relu)(inputs)
x = tf.keras.layers.Dense(128, activation=tf.nn.relu)(x)
outputs = tf.keras.layers.Dense(2, activation=tf.nn.softmax)(x)

tf_model = tf.keras.Model(inputs=inputs, outputs=outputs)
```

```python
tf_model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=['accuracy']
)
```

```python
tf_model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    batch_size=32,
    epochs=10
)
```

# Results

```python
rf_score = rf.score(X_test, y_test)
tf_score = tf_model.evaluate(X_test, y_test, verbose=False)
```

```python
print(f" Random Forest Model: {rf_score}")
print(f"Neural Network Model: {tf_score[1]}")
```

## Seems the Random Forest model performs better than the ANN model for this dataset
